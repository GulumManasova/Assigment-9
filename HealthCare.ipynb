## Introduction
Kyrgyzstan, located in Central Asia, faces various public health challenges,
especially regarding diseases such as tuberculosis, respiratory illnesses, 
and high maternal and infant mortality rates. This dataset provides valuable insights into these health issues,
offering data on doctors, nursing staff, disease statistics, and mortality rates over multiple years.

This project aims to analyze healthcare data from Kyrgyzstan to identify trends in healthcare resources and 
health outcomes. The dataset has been selected for its relevance to understanding the effectiveness of healthcare 
delivery in the region and its potential to inform policy and healthcare strategies.

## Data Description
The dataset used in this analysis was sourced from [name of the source, e.g., World Health Organization, Kyrgyzstan Health Ministry, etc.]. The data spans several years and includes various healthcare metrics, focusing on the health status of the Kyrgyz population. The key variables in the dataset are as follows:

Year: The year the data was recorded.

Regions: The geographical regions within Kyrgyzstan.

Doctors: The number of doctors available in each region.

Tuberculosis: The number of tuberculosis cases recorded.

Respiratory disease: The number of cases related to respiratory diseases.

Acute bowel: Cases of acute bowel diseases.

Nursing staff: The number of nurses working in the regions.

Maternal mortality: The maternal mortality rate (deaths per 100,000 live births).

Infant mortality: The infant mortality rate (deaths per 1,000 live births).

## Data Analysis
The exploratory data analysis (EDA) was performed to understand the distribution and relationships between key variables. Descriptive statistics revealed the following:

The Maternal mortality rate showed significant fluctuations over the years, with peaks in specific regions.

A high correlation was found between doctors and nursing staff, indicating a strong relationship between healthcare staffing levels.

Disease trends showed an increase in tuberculosis cases over the years, while respiratory diseases remained stable.

Additionally, a linear regression analysis was conducted to examine the relationship between doctors and maternal mortality, which revealed that the number of doctors had a weak negative correlation with maternal mortality, suggesting that more healthcare professionals may help reduce maternal deaths, but the effect is not very strong
